```python
import pandas as pd
import numpy as np

def read_csv_file(file_path):
    try:
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        return df
    except Exception as e:
        print("An error occurred:", e)
        return None

# File path of the CSV file
file_path = "../Data/Population_Data/2023_pop_percentage.csv" 

# Call the function to read the CSV file
percent_pop_df = read_csv_file(file_path)

# Print the DataFrame
if percent_pop_df is not None:
    print(percent_pop_df)
else:
    print("Failed to read the CSV file.")
```

```python
over_21 = percent_pop_df.drop(percent_pop_df.index[0:21])
over_21_df = pd.DataFrame(over_21)
over_21_df = over_21_df.astype({'Male % of Population': float, 'Female % of Population': float})
over_21_df
```

```python
population_2023 = 339996563
```

```python
# Loop through each age and multiply "Male % of Population" by "Population 2023"
for Age in over_21_df.index:
    over_21_df.loc[Age, "Male Population 2023"] = over_21_df.loc[Age, "Male % of Population"] * population_2023

# Loop through each age and multiply "Female % of Population" by "Population 2023"
for Age in over_21_df.index:
    over_21_df.loc[Age, "Female Population 2023"] = over_21_df.loc[Age, "Female % of Population"] * population_2023

# Display the modified DataFrame
over_21_df['Male Population 2023']=over_21_df['Male Population 2023'].astype(int)
over_21_df['Female Population 2023']=over_21_df['Female Population 2023'].astype(int)
over_21_df
```

```python
for Age in over_21_df.index:
    over_21_df.loc[Age, "Population by Age 2023"] = over_21_df.loc[Age, "Male Population 2023"] + over_21_df.loc[Age, "Female Population 2023"]

over_21_df
```

```python
legal_drinking_age_total = over_21_df['Population by Age 2023'].sum()
legal_drinking_age_total
```
